# Question: What were the top 5 reoccurring Genres on the spotify "Top Tracks" playlist of 2024?

Its going to be a retrospective study. even though this information is recent, its still information from the past that has been completed (no more information is being added to this data) I'm going to use a census. I need to separate the top 5 genres from all the other genres that made the leader board that year. With a census, I can isolate specific aspects of my data easier and I can get more accurate results. (I need to examine a population rather than a sample)

I'm using the data that the teachers gave us. however, with the data that she pulled, I'm only going to be pulling from a small pool of the entirety of the data. I'm not sure where the data comes from but I believe it comes directly from the spotify playlist from last year

The data came from real people and not AI. Since the Data is from real people, there are already a ton of biases. The biggest bias I see comes from the fact that the people who made this playlist all have different music tastes. While that's completely fine, for something that requires quantitative data, it is going to throw off the data. not to mention social media is also a huge influence on the songs people like, whether they want to admit it or not.

By looking at both the pie chart AND the bar graph, the Top 5 genres are: Pop, country, R&B, Hip-hop- and another form of pop

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
top_tracks = pd.read_csv('Tracks.csv')

# Combine the 'Wiki Genre' and 'AI Genre' columns into a single list of genres
# Drop NaNs first
wiki_genres = top_tracks['Wiki Genre'].dropna()
ai_genres = top_tracks['AI Genre'].dropna()

# Combine all genre entries into one Series
all_genres = pd.concat([wiki_genres, ai_genres])

# If genres are comma-separated lists, split them
# Flatten the list if there are multiple genres per entry
split_genres = all_genres.str.split(',').explode().str.strip()

# Count genre frequency
genre_counts = split_genres.value_counts()

# Get top 5 genres
top_5_genres = genre_counts.head(5)
print("Top 5 Genres Frequency Table:")
print(top_5_genres)

# Plot a pie chart of the top 5 genres
plt.figure(figsize=(8, 6))
colors = sns.color_palette('pastel')[0:5]
top_5_genres.plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black'}
)

plt.title('Top 5 Most Frequent Genres')
plt.ylabel('')  # Remove y-axis label for cleaner look
plt.tight_layout()
plt.show()

```
