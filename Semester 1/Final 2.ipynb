# Question: In Blackjack, suppose a player has a 1.92% chance of winning a single game. Assuming each game is independent, what is the probability that the player wins for the first time on their 7th game?
# answer: The Probability of the first win against the AI dealer would be a 1.51

```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
import time

# Create the Toy Text Blackjack environment
env = gym.make('Blackjack-v1', render_mode='human')
observation, info = env.reset()

# Simulation parameters
num_episodes = 100
steps_per_episode = 50
total_steps = 0

# Track results
wins = 0
losses = 0
draws = 0

# Run simulation (text-based)
for episode in range(num_episodes):
    total_reward = 0
    observation, info = env.reset()

    for step in range(steps_per_episode):
        action = env.action_space.sample()  # Random action: hit or stick
        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        total_steps += 1

        if terminated or truncated:
            break

    # Determine outcome
    if total_reward > 0:
        result = "✅ WIN"
        wins += 1
    elif total_reward < 0:
        result = "❌ LOSS"
        losses += 1
    else:
        result = "➖ DRAW"
        draws += 1

    print(f"Episode {episode+1}/{num_episodes}: Total reward = {total_reward:.0f} → {result}")

env.close()
print(f"\nSimulation complete — {num_episodes} episodes, {total_steps} total steps.")
print(f"Wins: {wins}, Losses: {losses}, Draws: {draws}\n")

# Visual Simulation (text-based rendering)
env = gym.make('Blackjack-v1', render_mode='human')
observation, info = env.reset()

fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')

# Reset counters for visualization
wins = losses = draws = 0

for episode in range(100):  # run 100 visual trials
    total_reward = 0
    observation, info = env.reset()

    for step in range(50): 
        action = env.action_space.sample()
        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

        # Text render (ANSI output)
        clear_output(wait=True)
        ax.clear()
        frame = env.render()  # returns text representation
        ax.text(0.01, 0.9, f"Episode {episode+1}/100 | Step {step+1} | Reward: {total_reward:.0f}", 
                fontsize=10, fontfamily="monospace", va='top')
        ax.text(0.01, 0.75, frame, fontsize=10, fontfamily="monospace", va='top')
        ax.axis('off')
        display(fig)

        time.sleep(0.25)

        if terminated or truncated:
            break

    # Determine outcome visually
    if total_reward > 0:
        wins += 1
        result = "✅ WIN"
    elif total_reward < 0:
        losses += 1
        result = "❌ LOSS"
    else:
        draws += 1
        result = "➖ DRAW"

    print(f"Episode {episode+1}: {result} (Reward: {total_reward:.0f})")

plt.close()
env.close()
print(f"\nVisual simulation complete — 100 episodes finished.")
print(f"Total Wins: {wins}, Losses: {losses}, Draws: {draws}")
```
