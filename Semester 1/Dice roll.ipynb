```python
#import section
import random,os,time


#play again
approved_Inputs = ['yes', 'no', 'n', 'y']
roll_Again = True
dice = [" "]*9
#Custom functions

def draw_Dice():
  global dice
  print(" -----")
  print(f'|{dice[0]} {dice[1]} {dice[2]}|')
  print(f'|{dice[3]} {dice[4]} {dice[5]}|')
  print(f'|{dice[6]} {dice[7]} {dice[8]}|')
  print(" -----")

def roll_Dice():
  global dice
  dice=[" "]*9
  dice_Number = random.randint(1,6)
  if dice_Number == 1:
    dice[4] = "O"
  elif dice_Number == 2:
    dice[0], dice[8] = "O"*2
  elif dice_Number == 3:
    dice[0], dice[4], dice[8] = "O"*3
  elif dice_Number == 4:
    dice[0], dice[2], dice[6], dice[8] = "O"*4
  elif dice_Number == 5:
    dice[0], dice[2], dice[4], dice[6], dice[8] = "O"*5
  else:
    dice[0], dice[2], dice[3], dice[5], dice[6], dice[8] = "O"*6
  
#main code
print("---Welcome to the dice roll program---")
while roll_Again == True:
  roll_Dice()
  draw_Dice()
  print("Would you like to roll again?")
  userInput = input()
  while userInput not in approved_Inputs:
    print("Please type in a valid input")
    userInput = input()
  if userInput.lower() == "n" or userInput.lower() == "no": 
    roll_Again = False
    print("Thanks for rolling the dice")
    time.sleep(1)
    
    
  elif userInput.lower() == "yes" or userInput.lower()=="y":
    roll_Again = True
  os.system("clear")
```
