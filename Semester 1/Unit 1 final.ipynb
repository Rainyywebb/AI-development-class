# Question: Within Amazons 50 bestselling books of 2019, what is the distribution of book prices, and which genres reoccurred the most to the bestseller list?
first for my question, its going to be an observational study since I wont be changing the outcome of the data I'm using nor will I be changing anything in the study. 
this study would be considered a sample. The entire population would be only books from 2019, however I am taking a sample out of the population (Im only taking the top 50 books out of the entire 2019 population)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

books = pd.read_csv('books.csv')

books_2019 = books[books['Year'] == 2019]

plt.figure(figsize=(8, 5))
sns.histplot(books_2019['Price'], bins=10, kde=True)
plt.title("Distribution of Book Prices (2019 Top 50 Bestsellers)")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.show()

mean_price = books_2019['Price'].mean()
median_price = books_2019['Price'].median()
mode_price = books_2019['Price'].mode().values[0]
std_price = books_2019['Price'].std()
min_price = books_2019['Price'].min()
max_price = books_2019['Price'].max()

print(f"Mean Price: ${mean_price:.2f}")
print(f"Median Price: ${median_price:.2f}")
print(f"Mode Price: ${mode_price:.2f}")
print(f"Standard Deviation: ${std_price:.2f}")
print(f"Price Range: ${min_price:.2f} - ${max_price:.2f}")

genre_counts = books_2019['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))


sns.barplot(data=genre_counts, x='Genre', y='Count', ax=ax1, palette="viridis")
ax1.set_title("Number of 2019 Bestsellers by Genre")
ax1.set_xlabel("Genre")
ax1.set_ylabel("Count of Books")
ax1.tick_params(axis='x', rotation=45)


ax2.axis('tight')
ax2.axis('off')
table = ax2.table(cellText=genre_counts.values,
                  colLabels=genre_counts.columns,
                  cellLoc='center',
                  loc='center')
table.scale(1.2, 1.2)
ax2.set_title("Frequency Table of Genres (2019)", fontsize=12, pad=10)

plt.tight_layout()
plt.show()
```
